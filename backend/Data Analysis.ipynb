```python
import numpy as np
import pandas as pd
```

## Song Dataset 

```python
songs_dataset = pd.read_csv('data/dataset.csv')
```

```python
songs_dataset.head()
```

```python
songs_dataset = songs_dataset[['track_name', 'artists', 'energy', 'loudness', 'mode', 'valence', 'track_genre', 'popularity']]
```

```python
songs_dataset.head()
```

```python
songs_dataset.shape
```

```python
songs = songs_dataset[songs_dataset['popularity'] >= 50]
```

```python
genre_list = songs_dataset['track_genre'].unique()
genre_list
```

```python
angry = ['alternative', 'black-metal', 'death-metal', 'dubstep', 'heavy-metal', 'grindcore', 'metal', 'grunge', 'industrial', 'punk']
disgust = ['children']
fear = ['ambient', 'chill', 'classical', 'deep-house', 'jazz', 'piano', 'study', 'sleep', 'trance', 'blues', 'guitar']
happy = ['acoustic', 'ambient', 'anime', 'classical', 'disco', 'disney', 'happy', 'jazz', 'j-pop', 'k-pop', 'reggae', 'guitar', 'gospel']
neutral = ['acoustic', 'hip-hop', 'jazz', 'pop', 'k-pop', 'rock', 'world-music', 'anime', 'indie', 'guitar', 'country']
sad = ['acoustic', 'alternative', 'blues', 'goth', 'romance', 'sleep', 'soul', 'classical', 'piano', 'british', 'emo']
energetic = ['alt-rock', 'anime', 'bluegrass', 'breakbeat', 'detroit-techno', 'disco', 'dubstep', 'edm', 'hip-hop', 'house', 'rock-n-roll']
```

## Angry 

```python
songs_angry = songs[songs['track_genre'].isin(angry)]
```

```python
songs_angry.head()
```

```python
songs_angry.shape
```

```python
songs_angry = songs_angry[(songs_angry['energy'] >= 0.69) & (songs_angry['loudness'] >= -7) & (songs_angry['valence'] <= 0.46)]
```

```python
songs_angry
```

```python
songs_angry = songs_angry.dropna()
angry_rec = songs_angry.sample(n=10, replace=False)
angry_rec = angry_rec.loc[:, ['track_name', 'artists']]

angry_rec
```
